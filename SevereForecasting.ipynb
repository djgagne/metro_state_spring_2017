```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.tile_providers import STAMEN_TERRAIN
output_notebook()
```

```python
storm_tracks = pd.read_csv("http://ensemble.ucar.edu/stormviewer/data/track_step_NCAR_mem1_20170330.csv")
```

```python
storm_tracks.columns
```

```python
storm_tracks.plot(x="UP_HELI_MAX_max", y="HAIL_MAXK1_max", kind="scatter")
```

```python
bound = 10000000
p = figure(x_range=(-bound, bound), y_range=(-bound, bound))
p.add_tile(STAMEN_TERRAIN)
p.circle(storm_tracks["Centroid_Lon"], storm_tracks["Centroid_Lat"])

show(p)
```
